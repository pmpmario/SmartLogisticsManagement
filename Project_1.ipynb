# SQL

```python
CREATE SCHEMA `logistics_project_1` ;


USE logistics_project_1;

-- 1. Courier Staff
CREATE TABLE IF NOT EXISTS courier_staff (
    courier_id VARCHAR(50) PRIMARY KEY,
    name VARCHAR(150) NOT NULL,
    rating DECIMAL(3,1),
    vehicle_type VARCHAR(50)
);

-- 2. Shipments (depends on courier_staff)
CREATE TABLE IF NOT EXISTS shipments (
    shipment_id VARCHAR(50) PRIMARY KEY,
    order_date DATE NOT NULL,
    origin VARCHAR(100) NOT NULL,
    destination VARCHAR(100) NOT NULL,
    weight DECIMAL(10,2) NOT NULL,
    courier_id VARCHAR(50),
    status VARCHAR(50) NOT NULL,
    delivery_date DATE NULL,
    CONSTRAINT fk_courier
        FOREIGN KEY (courier_id)
        REFERENCES courier_staff(courier_id)
        ON UPDATE CASCADE
        ON DELETE SET NULL
);

-- 3. Shipment Tracking (depends on shipments)
CREATE TABLE IF NOT EXISTS shipment_tracking (
    tracking_id INT AUTO_INCREMENT PRIMARY KEY,
    shipment_id VARCHAR(50) NOT NULL,
    status VARCHAR(50) NOT NULL,
    timestamp DATETIME NOT NULL,
    CONSTRAINT fk_shipment_tracking
        FOREIGN KEY (shipment_id)
        REFERENCES shipments(shipment_id)
        ON UPDATE CASCADE
        ON DELETE CASCADE
);

-- 4. Costs (depends on shipments)
CREATE TABLE IF NOT EXISTS costs (
    shipment_id VARCHAR(50) PRIMARY KEY,
    fuel_cost DECIMAL(15,2) NOT NULL DEFAULT 0,
    labor_cost DECIMAL(15,2) NOT NULL DEFAULT 0,
    misc_cost DECIMAL(15,2) NOT NULL DEFAULT 0,
    CONSTRAINT fk_costs_shipment
        FOREIGN KEY (shipment_id)
        REFERENCES shipments(shipment_id)
        ON UPDATE CASCADE
        ON DELETE CASCADE
);

-- 5. Routes
CREATE TABLE IF NOT EXISTS routes (
    route_id VARCHAR(50) PRIMARY KEY,
    origin VARCHAR(100) NOT NULL,
    destination VARCHAR(100) NOT NULL,
    distance_km DECIMAL(10,2) NOT NULL,
    avg_time_hours DECIMAL(5,2) NOT NULL
);

-- 6. Warehouses
CREATE TABLE IF NOT EXISTS warehouses (
    warehouse_id VARCHAR(50) PRIMARY KEY,
    city VARCHAR(100) NOT NULL,
    state VARCHAR(50) NOT NULL,
    capacity INT NOT NULL
);
```

# STREAMLIT

```python
import streamlit as st
import pandas as pd
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# Database Connection
password = quote_plus("Alys@003!")
engine = create_engine(f"mysql+pymysql://root:{password}@localhost/logistics_project_1")

@st.cache_data(ttl=300)
def load_data():
    shipments = pd.read_sql("SELECT * FROM shipments", engine)
    couriers = pd.read_sql("SELECT * FROM courier_staff", engine)
    costs = pd.read_sql("SELECT * FROM costs", engine)
    routes = pd.read_sql("SELECT * FROM routes", engine)
    shipment_tracking = pd.read_sql("SELECT * FROM shipment_tracking", engine)
    warehouses = pd.read_sql("SELECT * FROM warehouses", engine)
    return shipments, couriers, costs, routes, shipment_tracking, warehouses

# Load all DataFrames
shipments, couriers, costs, routes, shipment_tracking, warehouses = load_data()
shipments["order_date"] = pd.to_datetime(shipments["order_date"])
shipments["delivery_date"] = pd.to_datetime(shipments["delivery_date"])

st.set_page_config(page_title="Logistics Dashboard", layout="wide")
st.title("🚚 Logistics Intelligence Dashboard")

# Tabs
tab_a, tab_b, tab_c, tab_d = st.tabs(["🔎 Shipment Filters", "📊 Operational KPIs", "📈 Analytical Views","💡 Business Insights"])

with tab_a:
    st.subheader("🔎 Filter Shipments")

    
    col1, col2 = st.columns(2)
    with col1:
        status_filter = st.multiselect("Status", shipments["status"].unique(), max_selections=3)
        origin_filter = st.multiselect("Origin", shipments["origin"].unique(), max_selections=3)
        shipment_id_filter = st.text_input("Shipment ID")
    with col2:
        destination_filter = st.multiselect("Destination", shipments["destination"].unique(), max_selections=3)
        courier_filter = st.multiselect("Courier", couriers["name"].unique(), max_selections=3)
        date_range = st.date_input(
            "Order Date Range",
            [shipments["order_date"].min(), shipments["order_date"].max()]
        )

    
    filtered = shipments.copy()
    if status_filter:
        filtered = filtered[filtered["status"].isin(status_filter)]
    if origin_filter:
        filtered = filtered[filtered["origin"].isin(origin_filter)]
    if destination_filter:
        filtered = filtered[filtered["destination"].isin(destination_filter)]
    if courier_filter:
        valid_couriers = couriers[couriers["name"].isin(courier_filter)]["courier_id"]
        filtered = filtered[filtered["courier_id"].isin(valid_couriers)]
    if shipment_id_filter:
        filtered = filtered[filtered["shipment_id"].str.contains(shipment_id_filter, case=False, na=False)]
    if len(date_range) == 2:
        filtered = filtered[
            (filtered["order_date"] >= pd.to_datetime(date_range[0])) &
            (filtered["order_date"] <= pd.to_datetime(date_range[1]))
        ]

    st.dataframe(filtered, height=400) 

with tab_b:
    st.subheader("📊 Operational KPIs")

    total_shipments = len(shipments)
    delivered_pct = (shipments["status"] == "Delivered").mean() * 100
    cancelled_pct = (shipments["status"] == "Cancelled").mean() * 100

    # Only consider delivered shipments for avg delivery time
    delivered = shipments[shipments["status"] == "Delivered"]
    avg_delivery_time = (pd.to_datetime(delivered["delivery_date"]) - pd.to_datetime(delivered["order_date"])).dt.days.mean()

    # Calculate total operational cost
    # Merge shipments with costs to ensure we only sum existing shipments
    shipments_costs = shipments.merge(costs, on="shipment_id", how="left")
    shipments_costs["total_cost"] = shipments_costs[["fuel_cost", "labor_cost", "misc_cost"]].sum(axis=1)
    total_operational_cost = shipments_costs["total_cost"].sum()

    # Create 5 columns
    col1, col2, col3, col4, col5 = st.columns(5)
    col1.metric("Total Shipments", total_shipments)
    col2.metric("Delivered %", f"{delivered_pct:.2f}%")
    col3.metric("Cancelled %", f"{cancelled_pct:.2f}%")
    col4.metric("Avg Delivery Time (days)", f"{avg_delivery_time:.2f}")
    col5.metric("Total Operational Cost", f"${total_operational_cost:,.2f}")


# Tab C: Analytical Views
with tab_c:
    st.subheader("📈 Delivery Performance Insights")
    shipments_routes = shipments.merge(routes, on=["origin","destination"], how="left")
    shipments_routes["delivery_days"] = (shipments_routes["delivery_date"] - shipments_routes["order_date"]).dt.days

    # Average delivery time per route
    avg_delivery_route = shipments_routes.groupby(["origin","destination"])["delivery_days"].mean().reset_index()
    st.write("Average Delivery Time per Route")
    st.dataframe(avg_delivery_route)

    # Most delayed routes
    most_delayed = avg_delivery_route.sort_values("delivery_days", ascending=False).head(10)
    st.write("Most Delayed Routes")
    st.dataframe(most_delayed)

    # Delivery time vs distance
    st.write("Delivery Time vs Distance")
    delivery_distance_df = shipments_routes[["shipment_id","distance_km","delivery_days","status"]]
    status_grouped = delivery_distance_df.groupby("status")[["delivery_days","distance_km"]].sum()
    st.bar_chart(status_grouped)

   
    st.subheader("🚴 Courier Performance")
    courier_perf = shipments.merge(couriers, on="courier_id", how="left")
    courier_summary = courier_perf.groupby("name").agg(
        shipments_handled=("shipment_id","count"),
        on_time_pct=("status", lambda x: (x=="Delivered").mean()*100),
        avg_rating=("rating","mean")
    ).reset_index()
    st.write("Shipments handled per Courier")
    st.dataframe(courier_summary)
    st.bar_chart(courier_summary.set_index("name")["shipments_handled"])

   
    st.subheader("💰 Cost Analytics")
    shipments_costs = shipments.merge(costs, on="shipment_id", how="left")
    shipments_costs["total_cost"] = shipments_costs[["fuel_cost","labor_cost","misc_cost"]].sum(axis=1)
    st.write("Total Cost per Shipment")
    st.dataframe(shipments_costs[["shipment_id","total_cost"]])

    cost_per_route = shipments_costs.merge(routes, on=["origin","destination"], how="left").groupby(["origin","destination"])["total_cost"].sum().reset_index()
    st.write("Total Cost per Route")
    st.dataframe(cost_per_route)

    total_fuel = shipments_costs["fuel_cost"].sum()
    total_labor = shipments_costs["labor_cost"].sum()
    total_misc = shipments_costs["misc_cost"].sum()
    st.bar_chart(pd.DataFrame({
        "cost_type":["fuel","labor","misc"],
        "amount":[total_fuel,total_labor,total_misc]
    }).set_index("cost_type"))

    high_cost = shipments_costs.sort_values("total_cost", ascending=False).head(10)
    st.write("High-Cost Shipments")
    st.dataframe(high_cost[["shipment_id","total_cost"]])

    
    st.subheader("❌ Cancellation Analysis")
    cancel_origin = shipments.groupby("origin")["status"].apply(lambda x: (x=="Cancelled").mean()*100).reset_index()
    st.bar_chart(cancel_origin.set_index("origin")["status"])

    cancel_courier = shipments.merge(couriers, on="courier_id", how="left").groupby("name")["status"].apply(lambda x: (x=="Cancelled").mean()*100).reset_index()
    st.bar_chart(cancel_courier.set_index("name")["status"])

    cancelled = shipments[shipments["status"]=="Cancelled"].copy()
    cancelled["time_to_cancel_days"] = (cancelled["delivery_date"] - cancelled["order_date"]).dt.days.fillna(0)
    st.write("Time-to-Cancellation (days)")
    st.bar_chart(cancelled.set_index("shipment_id")["time_to_cancel_days"])

    
    st.subheader("🏭 Warehouse Insights")
    st.bar_chart(warehouses.set_index("city")["capacity"])

    warehouse_traffic = shipments.merge(warehouses, left_on="origin", right_on="city", how="left").groupby("city")["shipment_id"].count().reset_index()
    st.bar_chart(warehouse_traffic.set_index("city")["shipment_id"])


# Tab D: Potential Business Insights

with tab_d:
    st.subheader("💡 Potential Business Insights")

    # 1️.Routes with highest delays
    shipments_routes = shipments.merge(routes, on=["origin","destination"], how="left")
    shipments_routes["delivery_days"] = (shipments_routes["delivery_date"] - shipments_routes["order_date"]).dt.days
    route_delays = shipments_routes.groupby(["origin","destination"])["delivery_days"].mean().sort_values(ascending=False).reset_index()
    st.write("Routes with Highest Delays")
    st.dataframe(route_delays.head(10))

    # 2️.Couriers handling most shipments
    courier_counts = shipments.merge(couriers, on="courier_id", how="left").groupby("name")["shipment_id"].count().reset_index()
    courier_counts = courier_counts.sort_values("shipment_id", ascending=False)
    st.write("Couriers Handling Most Shipments")
    st.dataframe(courier_counts)

    # 3️.High-rated couriers delivering faster
    courier_perf = shipments.merge(couriers, on="courier_id", how="left")
    courier_perf["on_time_days"] = (courier_perf["delivery_date"] - courier_perf["order_date"]).dt.days
    courier_rating_speed = courier_perf.groupby("name").agg(avg_rating=("rating","mean"), avg_delivery_days=("on_time_days","mean")).reset_index()
    st.write("High-Rated Couriers vs Avg Delivery Time")
    st.dataframe(courier_rating_speed)

    # 4️.Most expensive shipments
    shipments_costs = shipments.merge(costs, on="shipment_id", how="left")
    shipments_costs["total_cost"] = shipments_costs[["fuel_cost","labor_cost","misc_cost"]].sum(axis=1)
    high_cost_shipments = shipments_costs.sort_values("total_cost", ascending=False)
    st.write("Most Expensive Shipments")
    st.dataframe(high_cost_shipments[["shipment_id","total_cost","weight"]].head(10))

    # 5️.Cost vs Weight
    st.write("Cost vs Weight Comparison")
    st.bar_chart(high_cost_shipments.set_index("shipment_id")[["weight","total_cost"]])

    # 6️.Cities with maximum cancellations
    cancel_city = shipments[shipments["status"]=="Cancelled"].groupby("origin")["shipment_id"].count().reset_index().sort_values("shipment_id", ascending=False)
    st.write("Cities with Maximum Cancellations")
    st.dataframe(cancel_city)

   # 7️.Routes underperforming relative to distance
    # Merge shipments_routes with costs to get total_cost
    shipments_routes_costs = shipments_routes.merge(costs, on="shipment_id", how="left")
    shipments_routes_costs["total_cost"] = shipments_routes_costs[["fuel_cost","labor_cost","misc_cost"]].sum(axis=1)
    shipments_routes_costs["cost_per_km"] = shipments_routes_costs["total_cost"] / shipments_routes_costs["distance_km"]

    underperforming_routes = shipments_routes_costs.groupby(["origin","destination"])["cost_per_km"]\
        .mean().sort_values(ascending=False).reset_index()

    st.write("Routes Underperforming Relative to Distance")
    st.dataframe(underperforming_routes.head(10))
```
